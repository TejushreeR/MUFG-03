```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering

# 📘 Step 1: Create a small sample dataset
data = {
    'X': [1, 2, 5, 8, 9],
    'Y': [1, 1, 5, 8, 9]
}
df = pd.DataFrame(data)

# 📊 Step 2: Plot the original data points
plt.scatter(df['X'], df['Y'], color='blue')
for i, txt in enumerate(df.index):
    plt.annotate(txt, (df['X'][i]+0.1, df['Y'][i]+0.1))
plt.title("Original Data Points")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()

# 🧬 Step 3: Generate Dendrogram
linked = linkage(df, method='single')

plt.figure(figsize=(8, 4))
dendrogram(linked,
           labels=df.index.tolist(),
           distance_sort='descending')
plt.title("Dendrogram")
plt.xlabel("Data Point Index")
plt.ylabel("Distance")
plt.show()

# 🧩 Step 4: Perform Agglomerative Clustering
model = AgglomerativeClustering(n_clusters=2, metric='euclidean', linkage='single')
df['Cluster'] = model.fit_predict(df)

# 🎨 Step 5: Plot clusters
colors = ['red', 'green']
plt.figure()
for cluster in df['Cluster'].unique():
    clustered_data = df[df['Cluster'] == cluster]
    plt.scatter(clustered_data['X'], clustered_data['Y'], label=f'Cluster {cluster}', color=colors[cluster])
plt.title("Clustered Data")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```
